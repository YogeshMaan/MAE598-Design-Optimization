```python
import numpy as np
from bayes_opt import BayesianOptimization

# Define the cost function (you can replace this with your own function)
def cost_function(x1, x2):
    return -1*((4 - 2.1*(x1**2) + ((x1**4)/3))*(x1**2) + (x1*x2) + (-4 + 4*(x2**2))*(x2**2))

# Set up the optimization space
space = {'x1': (-3.0, 3.0), 'x2': (-2.0, 2.0)} # Bounds for x

# Initialize the optimizer
opt = BayesianOptimization(
    f = cost_function,
    pbounds = space,
    random_state = 10,
    allow_duplicate_points=True)


opt.maximize(init_points = 0, n_iter = 200)


print("Best Parameter:", opt.max['params'])
print("Best Cost:", opt.max['target'])
```
